```
parent_dir = 'drive/My Drive/BTP/'

class Args():
  def __init__(self):
    self.phase = 'train'
    self.dataset = 'CUB_200_2011'

    self.iteration = 1000000  # Number of iterations
    self.decay_flag = True    # deay flag
    self.decay_iter = 500000  # decay epoch

    self.batch_size = 8       # batch size
    self.print_freq = 1000    # The number of image_print_freq
    self.save_freq = 2000    # The number of ckpt_save_freq

    self.lr = 0.0002          # The learning rate

    self.gan_type = 'gan'     # Choose from ('gan', 'lsgan', 'hinge')

    self.adv_weight = 1       # Weight about GAN
    self.kl_weight = 1        # Weight about kl_loss
    self.embed_weight = 1     # Weight about embed_weight

    self.z_dim = 100          # condition and noise z dimension
    self.embed_dim = 256      # embedding deminsion
    self.g_dim = 32           # generator feature basic dimension
    self.d_dim = 64           # discriminator feature basic dimesion

    self.sn = False           # using spectral norm

    self.img_height = 128     # The height of image
    self.img_width = 128      # The width of image
    self.img_ch = 3           # The size of image channel
    self.augment_flag = False # Image augmentation use or not

    self.checkpoint_dir = parent_dir + 'checkpoint'  # Directory name to save the checkpoints
    self.result_dir = parent_dir + 'results'         # Directory name to save the generated images
    self.log_dir = parent_dir + 'logs'               # Directory name to save training logs
    self.sample_dir = parent_dir + 'samples'         # Directory name to save the samples on training
```

```
import shutil

def setupClassInfo(args):
  folder = './dataset/' + args.dataset + '/text/'
  file = folder + 'class_info/class_info_' + args.phase + '.pickle'
  shutil.copyfile(file, folder + 'class_info.pickle')
```

```
from AttnGAN import AttnGAN
from utils import *

args = Args()
```

```
args.phase = 'train'
setupClassInfo(args)
gan = AttnGAN(args)
gan.build_model()

gan.train()
print(" [*] Training finished!")
```
